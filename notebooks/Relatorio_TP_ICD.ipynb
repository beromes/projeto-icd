# **Introdução**
Este relatório tem como objetivo apresentar resultados detalhados das relações entre a taxa de natalidade e o período pandêmico no Brasil. Para isso, foram utilizados os métodos de ciência de dados, com intuito de explorar, testar e correlacionar os dados disponíveis.

É importante ressaltar que o escopo dessa análise será baseado nas técnicas e metodologias utilizadas, portanto as conclusões devem ser consideradas dentro desse limite. Os resultados devem servir como incentivo para investigações mais profundas sobre o tema.


# **Motivação**
A pandemia de COVID-19 teve impacto mundial, gerando inúmeras consequências, que ainda estão sendo estudadas e analisadas com intuito de compreender quais foram os efeitos negativos para a sociedade e como eles podem ser contornados ou reduzidos.

Ao analisar a taxa de natalidade em relação ao período pandêmico no Brasil busca-se compreender como a pandemia de COVID-19 afetou a quantidade de nascidos, se houve aumento ou diminuição e quais as potenciais causas.


# **Perguntas**


* Qual foi o impacto da pandemia de COVID-19 na taxa de natalidade no Brasil e em suas regiões, considerando fatores socioeconômicos e políticas públicas adotadas durante o período pandêmico?

* Qual o comportamento da taxa de natalidade em diferentes regiões que tiveram mais e menos casos de COVID-19 no Brasil?

* Com base nos dados do período pré-pandemia, qual seria a tendência esperada da taxa de natalidade no Brasil nos anos seguintes, e como essa tendência foi afetada pela pandemia? É possível prever a recuperação da taxa de natalidade no Brasil nos próximos anos?

* É possível prever a taxa de natalidade nos próximos períodos pandêmicos no Brasil de acordo com os comportamentos observados?




# **Dados**

Ela cobre um período de tempo desde o ano de 1994 até 2021.

Com intuito de facilitar o acesso e manipulação dos dados, foi realizado um recorte em uma amostra de 100.000 registros aleatórios dentre os anos de 2015 e 2021.

## **Conclusão**

Apesar da hipótese nula ter sido rejeitada, com a análise exploratória e com a regressão linear realizada, percebemos que os nascidos não foram afetados significativamente pelo período pandêmico. A justificativa para o desvio no teste de hipótese está relacionada ao descrescimento que já ocorria ao longo dos anos no período pré pandêmico, juntamente com a falta dos dados pós pandêmicos.
